```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importando dataset

df_enem_2023 = pd.read_csv(r"C:\Users\Thomaz David\Desktop\Enem\MICRODADOS_ENEM_2023.csv", sep=";", encoding = 'ISO-8859-1')
```

```python
#Observando o dataset

df_enem_2023.columns
```

```python
#Filtração de dados 1.0

colunas_selecionadas_enem2023 = ['TP_FAIXA_ETARIA','NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_MT','TP_SEXO','TP_COR_RACA']
```

```python
#Filtração de dados 1.2
enem_2023_filtrado = df_enem_2023.filter(items=colunas_selecionadas_enem2023)
```

```python
enem_2023_filtrado_100 = enem_2023_filtrado[0:100]
```

```python
sexo_enem_2023 = enem_2023_filtrado['TP_SEXO']
```

```python
#Quantidade de pessoas por sexo no dataset Enem 2023


print(sexo_enem_2023.value_counts())

#Quantidade de pessoas por sexo no dataset Enem 2023

sexo_enem_2023.value_counts().plot(kind='bar', color=['blue', 'red'])
plt.title('Quantidade de pessoas por sexo no dataset Enem 2023')
plt.xlabel('Sexo')
plt.ylabel('Quantidade de pessoas em milhões')
plt.show()


```

```python
#Quantidade de pessoas por faixa etária no dataset Enem 2023

faixa_etaria_enem_2023 = enem_2023_filtrado['TP_FAIXA_ETARIA']

#Quantidade de pessoas por faixa etária no dataset Enem 2023

faixa_etaria_enem_2023.value_counts().plot(kind='bar', color=['blue', 'red'])
plt.title('Quantidade de pessoas por faixa etária no dataset Enem 2023')
plt.xlabel('Faixa etária')
plt.ylabel('Quantidade de pessoas')
plt.show()
```

```python
#FILTRANDO NOTAS POR AREA DE CONHECIMENTO

enem_nota_Cn = enem_2023_filtrado['NU_NOTA_CN']
enem_nota_Mt = enem_2023_filtrado['NU_NOTA_MT']
enem_nota_Ch = enem_2023_filtrado['NU_NOTA_CH']
enem_nota_Lc = enem_2023_filtrado['NU_NOTA_LC']

```
